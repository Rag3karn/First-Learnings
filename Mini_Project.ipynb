```python
#Group by age
young=np.sum((df['Age']>=0)&(df['Age']<20)) #0 to 19
adult=np.sum((df['Age']>=20)&(df['Age']<40)) #20 to 39
midage=np.sum((df['Age']>=40)&(df['Age']<60)) #40 to 59
old=np.sum(df['Age']>=60) #>=60
print(young,adult,midage,old)
```

```python
plt.plot(df['Age'],df["Annual Income (k$)"])
plt.title("AGE vs Annual income")
plt.xlabel('Age')
plt.ylabel('Annual Income (k$)')
plt.grid()
```

```python
plt.plot(df['Annual Income (k$)'],df['Spending Score (1-100)'])
plt.title("Annual income vs Spending Score")
plt.xlabel('Annual income')
plt.ylabel('Spending Score')
plt.grid()
```

```python
Annual_income_min=min(df['Annual Income (k$)'].values)
Annual_income_max=max(df['Annual Income (k$)'].values)
Age_min=min(df['Age'].values)
Age_max=max(df['Age'].values)
Annual_income_norm=(df['Annual Income (k$)'].values-Annual_income_min)/(Annual_income_max-Annual_income_min)
Age_norm=(df['Age'].values-Age_min)/(Age_max-Age_min)
plt.plot(Age_norm,Annual_income_norm,color='Red')
plt.xlabel('Age')
plt.ylabel('Annual Income')
plt.grid()
```
